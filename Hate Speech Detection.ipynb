```python
dataset["labels"]=dataset["class"].map({0:"Hate Speech",
                                      1:"Offensive Language",
                                      2:"No hate or offensive Language"}
                                     )
```

```python
import re
import nltk
nltk.download("stopwords")
import string
```

```python
# removal of stop words and stemming the words
from nltk.corpus import stopwords
stopwords=set(stopwords.words("english"))
```

```python
#import stemming
stemmer= nltk.SnowballStemmer("english")
```

```python
#data cleaning
def clean_data(text):
    text = str(text).lower()
    text = re.sub('https?://\S+|www.S+','',text)
    text = re.sub('\[.*?\]', '',text)
    text = re.sub('<.*?>+', '',text)
    text = re.sub('[%s]' %re.escape(string.punctuation), '',text)
    text = re.sub('\n', '',text)
    text = re.sub('\w*\d\w*', '',text)
    #stop words removal
    text = [word for word in text.split(' ') if word not in stopwords]
    text= " ".join(text)
    # Stemming the text
    text = [stemmer.stem(word) for word in text.split(" ")]
    text= " ".join(text)
    return text
```

```python
data["tweet"]=data["tweet"].apply(clean_data)
```

```python
data
```

```python
X = np.array(data["tweet"])
Y = np.array(data["labels"])
X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv=CountVectorizer()
X=cv.fit_transform(X)
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
# building out ml model
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
Y_pred= dt.predict(X_test)
```

```python
# confusion matrix and accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
cm
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'dataset' is your DataFrame

# Step 1: Identify non-numeric columns
non_numeric_columns = dataset.select_dtypes(include=['object']).columns
print("Non-numeric columns: ", non_numeric_columns)

# Inspect non-numeric columns to understand their content
for col in non_numeric_columns:
    print(f"Sample values in column {col}: ", dataset[col].unique()[:5])

# Step 2: Convert non-numeric columns if possible
# For simplicity, let's drop them
numeric_dataset = dataset.drop(columns=non_numeric_columns)

# Step 3: Check for any remaining non-numeric values within numeric columns
for col in numeric_dataset.columns:
    non_numeric_values = numeric_dataset[~numeric_dataset[col].apply(lambda x: np.issubdtype(type(x), np.number))]
    if not non_numeric_values.empty:
        print(f"Non-numeric values in column {col}: ", non_numeric_values)

# Step 4: Handle missing values, if any
# You can drop rows with missing values or fill them with some value (e.g., mean)
numeric_dataset = numeric_dataset.dropna()  # Dropping rows with any NaNs

# Ensure all data is numeric
numeric_dataset = numeric_dataset.apply(pd.to_numeric, errors='coerce')

# Drop any rows that couldn't be converted to numeric
numeric_dataset = numeric_dataset.dropna()

# Step 5: Compute the correlation matrix on the numeric data
correlation_matrix = numeric_dataset.corr()

# Step 6: Plot the correlation matrix
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
from wordcloud import WordCloud

def generate_wordcloud(data, title):
    wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(' '.join(data))
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Generate word clouds for each class
for class_label in dataset['class'].unique():
    class_data = dataset[dataset['class'] == class_label]['tweet']
    generate_wordcloud(class_data, f"Word Cloud for Class {class_label}")
```
